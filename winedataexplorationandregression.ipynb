```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn import metrics
import matplotlib.pyplot as plt

white2 = white
red2 = red

X = white2[['free sulfur dioxide', 'total sulfur dioxide', 'residual sugar']]
Y = white2['quality']

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)


model = LinearRegression()

# Fit the model to the training data
model.fit(X_train, y_train)

# Make predictions on the test set
Y_pred = model.predict(X_test)

#MSE for linear model
print('Mean Squared Error 1:', metrics.mean_squared_error(y_test, Y_pred))

plt.scatter(range(len(y_test)), y_test, color='brown', label='Actual')


plt.scatter(range(len(Y_pred)), Y_pred, color='black', label='Predicted')

plt.xlabel('Independent Variables')
plt.ylabel('Quality')
plt.title('Actual vs Predicted Quality')

# Adding a legend
plt.legend()

# Displaying the plot
plt.show()

# Initialize the decision tree regressor
regressor = DecisionTreeRegressor()

# Train the regressor on the training data
regressor.fit(X_train, y_train)
scores = cross_val_score(regressor, X, Y, cv=5, scoring='neg_mean_squared_error')

# Make predictions on the testing data
y_pred = regressor.predict(X_test)

# Calculate the mean squared error of the model
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
mse_mean = -np.mean(scores)
print (mse_mean)
```
